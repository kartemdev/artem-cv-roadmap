```python
import argparse

parser = argparse.ArgumentParser()

parser.add_argument('-n', '--name', required=True, help='your name', type=str)
parser.add_argument('-a', '--age', required=True, help='your age', type=int)
parser.add_argument('-c', '--country', default='Russia', choices=['USA', 'Europe', 'Asia'], help='choice country', type=str)
parser.add_argument('-is', '--is-student', action='store_true', default=False)
parser.add_argument('-s', '--skills', nargs='*', type=str, default=[])
```

```python
args = parser.parse_args('-n Artem -a 25 -s Python JS TS'.split())
print(args)
```
